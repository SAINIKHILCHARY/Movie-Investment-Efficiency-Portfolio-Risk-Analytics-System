Production companies need a data-driven approach to evaluate historical movie performance, identify investment patterns, reduce portfolio risk, and improve future investment decisions.

- Successfully profile and clean the movie dataset.
- Calculate financial metrics such as Profit, ROI, and Revenue-to-Budget Ratio.
- Identify profitable genres and production companies.
- Analyze relationships between Budget, Revenue, Profit, ROI, and Popularity.
- Build an interactive Power BI dashboard.
- Generate actionable business recommendations.
