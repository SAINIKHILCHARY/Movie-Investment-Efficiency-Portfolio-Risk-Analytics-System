```python
import pandas as pd

# Load dataset
df = pd.read_csv("project_dataset.csv")

# Create datatype report
datatype_report = pd.DataFrame({
    "Column_Name": df.columns,
    "Data_Type": df.dtypes.astype(str).values
})

print("DATA TYPE REPORT")
print("=" * 60)

print(datatype_report.to_string(index=False))

# Count columns by datatype
print("\nDATA TYPE SUMMARY")
print("=" * 60)

print(df.dtypes.value_counts())

# Export report
datatype_report.to_csv(
    "Day3_Data_Type_Report.csv",
    index=False
)

print("\nSaved: Day3_Data_Type_Report.csv")
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("project_dataset.csv")

# Calculate missing values
missing_report = pd.DataFrame({
    "Column_Name": df.columns,
    "Missing_Count": df.isnull().sum().values,
    "Missing_Percentage": (
        df.isnull().sum().values / len(df) * 100
    ).round(2)
})

# Sort highest missing values first
missing_report = missing_report.sort_values(
    by="Missing_Count",
    ascending=False
)

print("MISSING VALUE REPORT")
print("=" * 70)

print(missing_report.to_string(index=False))

# Summary
print("\nSUMMARY")
print("=" * 70)

print(
    "Total Missing Values:",
    df.isnull().sum().sum()
)

print(
    "Columns With Missing Values:",
    (df.isnull().sum() > 0).sum()
)

print(
    "Columns Without Missing Values:",
    (df.isnull().sum() == 0).sum()
)

# Export report
missing_report.to_csv(
    "Day3_Missing_Value_Report.csv",
    index=False
)

print("\nSaved: Day3_Missing_Value_Report.csv")
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("project_dataset.csv")

print("DUPLICATE ANALYSIS")
print("=" * 60)

# Check if movie_id exists
if "movie_id" in df.columns:

    # Duplicate movie IDs
    duplicate_id_mask = df.duplicated(
        subset=["movie_id"],
        keep=False
    )

    duplicate_ids = df[duplicate_id_mask].sort_values(
        by="movie_id"
    )

    # Exact duplicate rows
    exact_duplicates = df[df.duplicated(
        keep=False
    )]

    print(
        "Total Records:",
        len(df)
    )

    print(
        "Unique movie_id Values:",
        df["movie_id"].nunique()
    )

    print(
        "Missing movie_id Values:",
        df["movie_id"].isnull().sum()
    )

    print(
        "Rows Involved in Duplicate movie_id:",
        duplicate_id_mask.sum()
    )

    print(
        "Number of movie_id Values Repeated:",
        df.loc[
            duplicate_id_mask,
            "movie_id"
        ].nunique()
    )

    print(
        "Exact Duplicate Rows:",
        df.duplicated().sum()
    )

    # Display duplicates
    if len(duplicate_ids) > 0:

        print("\nDUPLICATE movie_id RECORDS")
        print("=" * 60)

        print(
            duplicate_ids[
                ["movie_id", "title"]
            ].to_string(index=False)
            if "title" in df.columns
            else duplicate_ids.to_string(index=False)
        )

        # Export duplicate records
        duplicate_ids.to_csv(
            "Day3_Duplicate_movie_id_Report.csv",
            index=False
        )

        print(
            "\nSaved: Day3_Duplicate_movie_id_Report.csv"
        )

    else:
        print("\nNo duplicate movie_id values found.")

else:
    print("ERROR: movie_id column not found.")
```
