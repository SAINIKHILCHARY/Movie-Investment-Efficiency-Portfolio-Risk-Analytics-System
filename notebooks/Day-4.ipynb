- Detect data-type issues (numeric-as-text, malformed dates in Release_Date, Release_Year)
- Detect invalid values / out-of-range in Revenue, Genre_Count, Production_Company_Count
- Detect inconsistent categories in Production_Companies, Primary_Production_Company, Cast_Gender_Ratio
- Write findings into profiling report

```python
import pandas as pd
import numpy as np

# Load Dataset
df = pd.read_csv("project_dataset.csv")

print("=" * 80)
print("TASK 1: DATA-TYPE ISSUE DETECTION")
print("=" * 80)

# Check current data types
print("\nCurrent Data Types:")
print(df.dtypes)

# Columns expected to be numeric
numeric_columns = [
    "Revenue",
    "Genre_Count",
    "Production_Company_Count",
    "Release_Year",
    "Cast_Gender_Ratio"
]

print("\nNUMERIC-AS-TEXT CHECK")
print("-" * 80)

for col in numeric_columns:

    converted = pd.to_numeric(df[col], errors="coerce")

    # Existing values that fail numeric conversion
    invalid_mask = df[col].notna() & converted.isna()

    print(f"\nColumn: {col}")
    print("Current Data Type:", df[col].dtype)
    print("Non-Numeric / Malformed Values:", invalid_mask.sum())

    if invalid_mask.sum() > 0:
        print(df.loc[invalid_mask, [col]].head(20))


# ---------------------------------------------
# Release_Date Validation
# ---------------------------------------------

parsed_dates = pd.to_datetime(
    df["Release_Date"],
    format="%d-%m-%Y %H:%M",
    errors="coerce"
)

invalid_date_mask = (
    df["Release_Date"].notna()
    & parsed_dates.isna()
)

print("\nRELEASE_DATE VALIDATION")
print("-" * 80)

print(
    "Malformed Release_Date Values:",
    invalid_date_mask.sum()
)

print(
    "Missing Release_Date Values:",
    df["Release_Date"].isna().sum()
)

if invalid_date_mask.sum() > 0:

    print("\nMalformed Date Records:")

    print(
        df.loc[
            invalid_date_mask,
            ["movie_id", "title", "Release_Date"]
        ].head(20)
    )


# ---------------------------------------------
# Release_Year Validation
# ---------------------------------------------

release_year = pd.to_numeric(
    df["Release_Year"],
    errors="coerce"
)

invalid_year_mask = (
    release_year.notna()
    & (
        (release_year < 1800)
        | (release_year > 2026)
    )
)

print("\nRELEASE_YEAR VALIDATION")
print("-" * 80)

print(
    "Invalid Release_Year Values:",
    invalid_year_mask.sum()
)

print(
    "Missing Release_Year Values:",
    release_year.isna().sum()
)


# ---------------------------------------------
# Release_Date vs Release_Year
# ---------------------------------------------

year_mismatch = (
    parsed_dates.notna()
    & release_year.notna()
    & (parsed_dates.dt.year != release_year)
)

print(
    "Release_Date / Release_Year Mismatches:",
    year_mismatch.sum()
)

if year_mismatch.sum() > 0:

    print(
        df.loc[
            year_mismatch,
            [
                "movie_id",
                "title",
                "Release_Date",
                "Release_Year"
            ]
        ].head(20)
    )
```

```python
import pandas as pd

df = pd.read_csv("project_dataset.csv")

print("=" * 80)
print("TASK 2: INVALID / OUT-OF-RANGE VALUE DETECTION")
print("=" * 80)


# =============================================
# Revenue
# =============================================

revenue = pd.to_numeric(
    df["Revenue"],
    errors="coerce"
)

print("\nREVENUE VALIDATION")
print("-" * 80)

print("Negative Revenue:", (revenue < 0).sum())

print("Zero Revenue:", (revenue == 0).sum())

print("Missing Revenue:", revenue.isna().sum())

print("Minimum Revenue:", revenue.min())

print("Maximum Revenue:", revenue.max())


# =============================================
# Genre_Count
# =============================================

genre_count = pd.to_numeric(
    df["Genre_Count"],
    errors="coerce"
)

print("\nGENRE_COUNT VALIDATION")
print("-" * 80)

print(
    "Negative Genre_Count:",
    (genre_count < 0).sum()
)

print(
    "Zero Genre_Count:",
    (genre_count == 0).sum()
)

print(
    "Missing Genre_Count:",
    genre_count.isna().sum()
)

print(
    "Minimum Genre_Count:",
    genre_count.min()
)

print(
    "Maximum Genre_Count:",
    genre_count.max()
)

print("\nGenre_Count Distribution:")

print(
    genre_count
    .value_counts(dropna=False)
    .sort_index()
)


# =============================================
# Production_Company_Count
# =============================================

company_count = pd.to_numeric(
    df["Production_Company_Count"],
    errors="coerce"
)

print("\nPRODUCTION_COMPANY_COUNT VALIDATION")
print("-" * 80)

print(
    "Negative Production_Company_Count:",
    (company_count < 0).sum()
)

print(
    "Zero Production_Company_Count:",
    (company_count == 0).sum()
)

print(
    "Missing Production_Company_Count:",
    company_count.isna().sum()
)

print(
    "Minimum Production_Company_Count:",
    company_count.min()
)

print(
    "Maximum Production_Company_Count:",
    company_count.max()
)

print("\nProduction_Company_Count Distribution:")

print(
    company_count
    .value_counts(dropna=False)
    .sort_index()
)
```

```python
import pandas as pd

df = pd.read_csv("project_dataset.csv")

print("=" * 80)
print("TASK 3: CATEGORY CONSISTENCY CHECK")
print("=" * 80)

categorical_columns = [
    "Production_Companies",
    "Primary_Production_Company"
]

for col in categorical_columns:

    print("\n" + "=" * 80)
    print("COLUMN:", col)
    print("=" * 80)

    text = df[col].astype("string")

    # Missing values
    print(
        "Missing Values:",
        df[col].isna().sum()
    )

    # Unique values
    print(
        "Unique Values:",
        df[col].nunique(dropna=True)
    )

    # Leading/trailing spaces
    whitespace_mask = (
        text.notna()
        & (text != text.str.strip())
    )

    print(
        "Leading/Trailing Space Issues:",
        whitespace_mask.sum()
    )

    # Blank strings
    blank_mask = (
        text.notna()
        & text.str.strip().eq("")
    )

    print(
        "Blank Strings:",
        blank_mask.sum()
    )

    # Placeholder values
    placeholders = [
        "unknown",
        "n/a",
        "na",
        "none",
        "null",
        "-"
    ]

    placeholder_mask = (
        text.str.strip()
        .str.lower()
        .isin(placeholders)
    )

    print(
        "Placeholder Values:",
        placeholder_mask.sum()
    )

    # Top values
    print("\nTop 20 Values:")

    print(
        df[col]
        .value_counts(dropna=False)
        .head(20)
    )


# =============================================
# Cast_Gender_Ratio Validation
# =============================================

print("\n" + "=" * 80)
print("CAST_GENDER_RATIO VALIDATION")
print("=" * 80)

cast_ratio = pd.to_numeric(
    df["Cast_Gender_Ratio"],
    errors="coerce"
)

print(
    "Missing Values:",
    cast_ratio.isna().sum()
)

print(
    "Negative Values:",
    (cast_ratio < 0).sum()
)

print(
    "Zero Values:",
    (cast_ratio == 0).sum()
)

print(
    "Minimum:",
    cast_ratio.min()
)

print(
    "Maximum:",
    cast_ratio.max()
)

print(
    "Median:",
    cast_ratio.median()
)
```

```python
import pandas as pd

df = pd.read_csv("project_dataset.csv")

findings = []


# Function to add findings
def add_finding(
    column,
    issue_type,
    issue_found,
    affected_rows,
    recommended_action
):

    findings.append({
        "Column": column,
        "Issue_Type": issue_type,
        "Issue_Found": issue_found,
        "Affected_Rows": int(affected_rows),
        "Recommended_Action": recommended_action
    })


# =============================================
# Release Date
# =============================================

parsed_dates = pd.to_datetime(
    df["Release_Date"],
    format="%d-%m-%Y %H:%M",
    errors="coerce"
)

malformed_dates = (
    df["Release_Date"].notna()
    & parsed_dates.isna()
)

add_finding(
    "Release_Date",
    "Data Type / Format",
    "Malformed date values",
    malformed_dates.sum(),
    "Review and standardize date format"
)


# =============================================
# Release Year
# =============================================

release_year = pd.to_numeric(
    df["Release_Year"],
    errors="coerce"
)

invalid_year = (
    release_year.notna()
    & (
        (release_year < 1800)
        | (release_year > 2026)
    )
)

add_finding(
    "Release_Year",
    "Range Check",
    "Invalid release year",
    invalid_year.sum(),
    "Validate against Release_Date"
)


# =============================================
# Revenue
# =============================================

revenue = pd.to_numeric(
    df["Revenue"],
    errors="coerce"
)

add_finding(
    "Revenue",
    "Invalid Value",
    "Negative revenue",
    (revenue < 0).sum(),
    "Investigate negative values"
)

add_finding(
    "Revenue",
    "Data Availability",
    "Zero revenue",
    (revenue == 0).sum(),
    "Determine whether zero means missing or genuine zero"
)


# =============================================
# Genre Count
# =============================================

genre_count = pd.to_numeric(
    df["Genre_Count"],
    errors="coerce"
)

add_finding(
    "Genre_Count",
    "Invalid Value",
    "Negative genre count",
    (genre_count < 0).sum(),
    "Investigate invalid values"
)

add_finding(
    "Genre_Count",
    "Data Availability",
    "Zero genre count",
    (genre_count == 0).sum(),
    "Compare with Genres column"
)


# =============================================
# Production Company Count
# =============================================

company_count = pd.to_numeric(
    df["Production_Company_Count"],
    errors="coerce"
)

add_finding(
    "Production_Company_Count",
    "Invalid Value",
    "Negative production company count",
    (company_count < 0).sum(),
    "Investigate invalid values"
)

add_finding(
    "Production_Company_Count",
    "Data Availability",
    "Zero production company count",
    (company_count == 0).sum(),
    "Compare with Production_Companies"
)


# =============================================
# Category Consistency
# =============================================

categorical_columns = [
    "Production_Companies",
    "Primary_Production_Company"
]

for col in categorical_columns:

    text = df[col].astype("string")

    whitespace = (
        text.notna()
        & (text != text.str.strip())
    )

    add_finding(
        col,
        "Category Consistency",
        "Leading/trailing whitespace",
        whitespace.sum(),
        "Standardize formatting during cleaning"
    )


# =============================================
# Cast Gender Ratio
# =============================================

cast_ratio = pd.to_numeric(
    df["Cast_Gender_Ratio"],
    errors="coerce"
)

add_finding(
    "Cast_Gender_Ratio",
    "Missing Values",
    "Missing gender ratio",
    cast_ratio.isna().sum(),
    "Investigate using male and female cast counts"
)

add_finding(
    "Cast_Gender_Ratio",
    "Invalid Value",
    "Negative gender ratio",
    (cast_ratio < 0).sum(),
    "Investigate invalid values"
)


# =============================================
# Create Report
# =============================================

profiling_report = pd.DataFrame(findings)

print("=" * 100)
print("DAY 4 DATA PROFILING REPORT")
print("=" * 100)

print(
    profiling_report.to_string(
        index=False
    )
)


# Export CSV
profiling_report.to_csv(
    "Day4_Data_Profiling_Report.csv",
    index=False
)

print(
    "\nSaved: Day4_Data_Profiling_Report.csv"
)
```
